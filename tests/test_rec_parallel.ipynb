```python
from loguru import logger

from pettingzoo.mpe import simple_push_v3
from pzu.wrappers import RecordVideo
```

```python
env = simple_push_v3.parallel_env(render_mode="rgb_array")
```

```python
max_episodes = 300
counter = 0

env = RecordVideo(env, "exports/rec_parallel_test")
obs, infos = env.reset()

while counter <= max_episodes:
    actions = {agent: env.action_space(agent).sample() for agent in obs.keys()}

    obs, rewards, terms, truncs, infos = env.step(actions)
    
    if any(terms.values()) or any(truncs.values()):
        counter += 1
        obs, infos = env.reset()
```

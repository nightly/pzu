```python
env = MarkovStagHunt(grid_size=(5, 5), render_mode="rgb_array", max_moves=100, 
                        one_hot_observations=True, flatten_observations=True)
```

```python
max_episodes = 300
counter = 0

env = RecordVideo(env, "exports/rec_parallel_test")
obs, infos = env.reset()

while counter <= max_episodes:
    actions = {agent: env.action_space(agent).sample() for agent in obs.keys()}

    obs, rewards, terms, truncs, infos = env.step(actions)
    
    if any(terms.values()) or any(truncs.values()):
        counter += 1
        obs, infos = env.reset()
```

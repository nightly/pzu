```python
from loguru import logger

from mpe2 import simple_tag_v3
from pzu.wrappers import RecordVideoAEC
```

```python
env = simple_tag_v3.env(render_mode="rgb_array", max_cycles=25)
```

```python
env = RecordVideoAEC(env, 
                  "exports/rec_aec_test",
                #   step_trigger=lambda x: x % 100 == 0
                )

for ep in range(100):
        env.reset(seed=42 + ep)
        for agent in env.agent_iter():
                obs, reward, terminated, truncated, info = env.last()
                action = None if (terminated or truncated) else env.action_space(agent).sample()
                env.step(action)

env.close()
```
